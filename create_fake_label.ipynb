```python

import pandas as pd
import random
```

```python
train = pd.read_csv('Data/train.csv')
test = pd.read_csv('Data/test.csv')
dev = pd.read_csv('Data/dev.csv')
```

```python
vowels = [
    'a ă â e ê i o ô ơ u ư y'.split(),
    'á ắ ấ é ế í ó ố ớ ú ứ ý'.split(),
    'à ằ ầ è ề ì ò ồ ờ ù ừ ỳ'.split(),
    'ả ẳ ẩ ẻ ể ỉ ỏ ổ ở ủ ử ỷ'.split(),
    'ã ẵ ẫ ẽ ễ ĩ õ ỗ ỡ ũ ữ ỹ'.split(),
    'ạ ặ ậ ẹ ệ ị ọ ộ ợ ụ ự ỵ'.split()
]
```

```python
set_word = set(' '.join(list(train.text)).split())
set_word
```

```python
dict_replace = dict()
for word in set_word:
    for vs in vowels:
        for ic, char in enumerate(vs):
            if char in word:
                dict_replace[word] = [word.replace(char, v[ic]) for v in vowels if char != v[ic] and word.replace(char, v[ic]) in set_word]
                if len(dict_replace[word]) == 0:
                    del dict_replace[word]

dict_replace
    
```

```python
len(dict_replace)
```

```python
def random_replace(string: str):
    string_ = string.split()
    string = string.split()
    index_replace = random.sample(list(range(len(string))), len(string) // 10 + 1)
    for i in index_replace:
        if string[i] in dict_replace:
            string[i] = random.choice(dict_replace[string[i]])
    return ' '.join(string), [int(s == s_) for s, s_ in zip(string, string_)]
```

```python
random_replace("tôi là ai tôi là ai tôi là ai tôi là ai")
```

```python
text_replace, score = zip(*list(train.text.apply(random_replace)))
score
```

```python
score = [' '.join([str(i) for i in s]) for s in score]
score
```

```python
train['score'] = score
train['text_replace'] = text_replace
```
